### Separar en conjuntos de train y test estratificados por 'Mother tongue' y 'Level'

```python
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
```

Mezclar las filas del dataset

```python
dfshuffle = shuffle(df3ppLimpio)
```

Separar el dataset original en variables por un lado (X) y clases por otro (y)

```python
X=dfshuffle.drop(['Mother tongue','Level'], axis=1)
```

```python
X
```

```python
y=dfshuffle[['Mother tongue','Level']]
```

```python
y
```

Obtener conjunto de train y test para X y para y, de tal manera que el 80% para entrenar y 20% para testear, y que esté estratificado por las dos variables de y

```python
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.20, random_state=42, stratify=y)
```

Comprobar que la estratificación en ambos subconjuntos está bien y es coherente.


Calculo primero el porcentaje en el dataset original

```python
df3ppLimpio.groupby(by=['Mother tongue','Level'])['Mother tongue'].count() /df3ppLimpio.shape[0]
```

Calculo en el ddataset de entrenamiento para ello, concateno el X e y

```python
train_set=pd.concat([X_train,y_train], axis=1)
```

```python
train_set.to_csv('train_set.csv', index=False, sep='\t')
```

```python
train_set.groupby(by=['Mother tongue','Level'])['Mother tongue'].count() /train_set.shape[0]
```

Hago lo mismo para el conjunto de test

```python
test_set=pd.concat([X_test,y_test], axis=1)
```

```python
test_set.to_csv('test_set.csv', index=False, sep='\t')
```

```python
aux=test_set.groupby(by=['Mother tongue','Level'])['Mother tongue'].count() /test_set.shape[0]
```

```python
aux.unstack(level='Mother tongue')
```

```python
aux.to_frame()
```

Comprobar que se han guardado bien:

```python
train_set = pd.read_csv('train_set.csv', sep='\t', encoding='utf-8') 
```

```python
train_set.info()
```

Separar el conjunto de train_set en train (90%) y validacion (10%)

```python
Xt=train_set.drop(['Mother tongue','Level'], axis=1)
```

```python
Xt
```

```python
yt=train_set[['Mother tongue','Level']]
```

```python
yt
```

```python
Xt_train, Xt_test, yt_train, yt_test=train_test_split(Xt,yt, test_size=0.10, random_state=42, stratify=yt)
```

```python
train_sett=pd.concat([Xt_train,yt_train], axis=1)
```

```python
train_sett.to_csv('train_train_set.csv', index=False, sep='\t')
```

```python
train_sett.groupby(by=['Mother tongue','Level'])['Mother tongue'].count() /train_sett.shape[0]
```

```python
validation_set=pd.concat([Xt_test,yt_test], axis=1)
```

```python
validation_set.to_csv('validation_train_set.csv', index=False, sep='\t')
```

```python
validation_set.groupby(by=['Mother tongue','Level'])['Mother tongue'].count() /validation_set.shape[0]
```
