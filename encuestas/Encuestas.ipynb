```python
import pandas as pd
import csv
import numpy as np
```

```python
df = pd.read_csv('test_set.csv', sep='\t', encoding='utf-8') 
```

```python
df.info()
```

```python
size = 2        # sample size
replace = False  # with replacement
fn = lambda obj: obj.loc[np.random.choice(obj.index, size, replace),:]
sentences=df.groupby(['Mother tongue','Level'], as_index=False).apply(fn)
```

```python
sentences.to_csv('train_set.csv', index=False, sep='\t')
```

```python
sentences.to_csv('sentencesRandom.csv', index=False, sep='\t')
```

```python
size = 4        # sample size
replace = False  # with replacement
fn = lambda obj: obj.loc[np.random.choice(obj.index, size, replace),:]
sentences20=sentences.groupby(['Level'], as_index=False).apply(fn)
```
