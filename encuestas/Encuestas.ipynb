```python
import pandas as pd
import csv
import numpy as np
```

### Cargar el conjunto de test previamente generado

```python
df = pd.read_csv('Conjuntos de datos/test_set.csv', sep='\t', encoding='utf-8') 
```

```python
df.info()
```

### Elegir una frase por lengua y nivel: en total 30 frases (6 lenguas x 5 niveles)

```python
size = 1        # sample size
replace = False  # with replacement
fn = lambda obj: obj.loc[np.random.choice(obj.index, size, replace),:]
sentences=df.groupby(['Mother tongue','Level'], as_index=False).apply(fn)
```

```python
sentences
```

### Eliminar del dataset original, las 30 frases elegidas, para completar con 10 más, 2 de cada nivel de manera aleatoria

```python
df_sentences=pd.merge(df, sentences, how='outer', indicator=True)
```

```python
df_sentences=df_sentences.loc[df_sentences._merge == 'left_only', ['Number','Word','Sentence','PpW2','PpW3','PpW4','PpW5','PpKenlm','Mother tongue','Level']]
```

```python
df_sentences
```

```python
size = 2        # sample size
replace = False  # with replacement
fn = lambda obj: obj.loc[np.random.choice(obj.index, size, replace),:]
aux_sentences=df_sentences.groupby(['Level'], as_index=False).apply(fn)
```

```python
aux_sentences
```

```python
dfsentences=pd.concat([sentences,aux_sentences], ignore_index=True)
```

```python
dfsentences
```

### Guardamos el datos generados en un archivo

```python
dfsentences.to_csv('sentences40Random.csv', index=False, sep='\t')
```

### Removemos de manera aleatoria los datos y los almacenamos en otro archivo

```python
from sklearn.utils import shuffle
sentences_shuffle = shuffle(dfsentences)
```

```python
sentences_shuffle.to_csv('sentences40RandomShuffle.csv', index=False, sep='\t')
```
