### Cargar el archivo con las frases ya elegidas

```python
df = pd.read_csv('sentences40RandomShuffle.csv', sep='\t', encoding='utf-8',index_col=0) 
```

### Declarar el número de personas, frases, ...

```python
n=4 #número personas
m=3 #número de veces que se repite cada frase 
f=40 #número de frases del dataset seleccionado
t=30 #número de frases por persona
```

### Generar un Dataframe dónde se va a distribuir el número de frase que se asigna a cada una de las personas (cada columna representa una persona)

```python
df_esquema = pd.DataFrame()
```

```python
array = np.empty((t, n), int)
df_esquema = pd.DataFrame(array)
```

```python
v=0 #contador de veces
s=0 #contador de frase
for x in range(0,t):
    for y in range(0, n):
            df_esquema.loc[x][y]=df.iloc[s]['Number']
            #print(df.iloc[s]['Number'])
            v=v+1
            if v==3:
                s=s+1
                v=0

        
```

```python
df_esquema
```

```python
for index in range(df_esquema.shape[1]):
    print('Frases para la persona: ', index)
    print('Número de las frases: ', df_esquema.iloc[: , index].values)
```

### Generar un diccionario con el mismo número de dataframes que de personas y guardar toda la información de la frase asignada anteriormente

```python
frames={}
groups=[1,2,3,4]
for grp in groups:
    df_aux=pd.DataFrame(columns=['Number','Word','Sentence','PpW2','PpW3','PpW4','PpW5','PpKenlm','Mother tongue','Level'])
    ar=df_esquema.iloc[: , gr].values
    for x in ar:
        df_aux=df_aux.append(df.loc[df['Number'] == x])
    frames[grp]=df_aux
```

```python
frames
```

### Guardar cada dataframe con la información completa en un csv

```python
#generar los csv con todos los datos
for grp in groups:
    frames[grp].to_csv("groupSentences_"+str(grp)+".csv", index=False, sep='\t')
```

### Eliminar todas las columnas menos "Sentence" y guardar en archivos csv para pasárselos a Mapi

```python
#dejar sólo las frases para pasarle a Mapi
for grp in groups:
    frames[grp]=frames[grp].drop(['Number','Word','PpW2','PpW3','PpW4','PpW5','PpKenlm','Mother tongue','Level'], axis=1)
```
